# Naive KNN in 2 Dimensions
### In this naiva approach, there is no pre processing. Run time is O(n) for a query point

```python
def getDistance(x1, y1, x2, y2):
    return ((x2-x1)**2 + (y2-y1)**2)**0.5


def knn(qx, qy, x, y, k=1):
    minDistance = 9999999;
    candidateX = None
    candidateY = None
    candidates = dict()
    for i in range(0, len(x)):
        dist = getDistance(x[i], y[i], qx, qy)
        if dist<minDistance:
            if len(candidates)<k:
                candidates["{},{}".format(x[i], y[i])] = dist
                
            else:
                # check if distance is less than the min distance in candidates
                localMaxDist = max(candidates.values())
                if dist < localMaxDist:
                    localMaxDistIndex = candidates.values().index(localMaxDist)
                    localMaxDistKey = candidates.items()[localMaxDistIndex][0]
                    candidates.pop(localMaxDistKey)
                    candidates["{},{}".format(x[i], y[i])] = dist

    retLst = list()

    for k in candidates.keys():
        xs,ys = k.split(',')
        retLst.append([int(xs), int(ys)])
    
    return retLst
```

```python
import random
n=1000
arr = np.random.randint(n, size=(2,n/2))
plt.scatter(arr[0], arr[1])
plt.xlim(0, n)
plt.ylim(0, n)

qp = [random.randint(0,n), random.randint(0,n)]
plt.scatter(qp[0], qp[1], color='red')

nn = knn(qp[0], qp[1], arr[0],arr[1], k=100)

for p in nn:
    plt.scatter(p[0],p[1], color='green')
```
